# comparative a greater percentage of people in senior citizen category have churned

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming `df` is your DataFrame
# Calculate the counts and normalize to percentages
counts = pd.crosstab(df['SeniorCitizen'], df['Churn'])
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot
fig, ax = plt.subplots(figsize=(3, 3))
bottom_values = [0] * len(percentages)

for churn_value, churn_percent in percentages.items():
    ax.bar(
        percentages.index, 
        churn_percent, 
        bottom=bottom_values, 
        label=f"Churn: {churn_value}"
    )
    # Update bottom values for stacking
    bottom_values += churn_percent

# Add annotations
for idx, (senior, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for churn_value, percent in row.items():
        ax.text(
            idx, 
            cumulative + percent / 2, 
            f"{percent:.1f}%", 
            ha='center', 
            va='center', 
            color='white', 
            fontsize=10
        )
        cumulative += percent

# Add labels and legend
plt.title("Churn by SeniorCitizen (Stacked Bar Chart)")
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")
plt.tight_layout()
plt.show()
```

```python
# he plots reveal that customers without optional services like Online Security, Online Backup, and Tech Support tend to churn more, indicating their importance for retention. Fiber optic internet customers exhibit the highest churn, suggesting potential dissatisfaction with the service. Overall, customers who subscribe to multiple services generally show lower churn rates.
```

```python

# Columns to plot
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Determine the layout of the subplots
num_columns = 3  # Number of columns in the grid
num_rows = -(-len(columns) // num_columns)  # Calculate rows (ceil division)

# Create the figure and axes
fig, axes = plt.subplots(num_rows, num_columns, figsize=(20, 5 * num_rows))

# Flatten axes for easy iteration
axes = axes.flatten()

# Create count plots
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i],hue=(df["Churn"]))
    axes[i].set_title(f"Count of {column}")
    axes[i].set_xlabel(column)
    axes[i].set_ylabel("Count")

# Remove empty subplots (if any)
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```
